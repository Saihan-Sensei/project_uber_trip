## TEMPORAL ANALYSIS

### By Hour

```python
df_hour_grouped = df.groupby(['Hour']).count() 
```

```python
df_hour = pd.DataFrame({'No.of.trips':df_hour_grouped.values[:,0]}, index = df_hour_grouped.index)
```

```python
df_hour.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))  
plt.plot(df_hour.index, df_hour['No.of.trips'], marker='o', linestyle='-', color='b')

plt.title('Number of Trips by Hour', fontsize=16)
plt.xlabel('Hours', fontsize=12)
plt.ylabel('Number of Trips', fontsize=12)

plt.grid(True)
plt.show()
```

 At 5:00 PM, it shows highest demand : possible reasons
 - Rush Hour Traffic, Evening Activities, End of work-day, etc.

Around 12-5 AM, it shows least number of trips, due to night-time

There is gradual increment in trips from 6AM

## Geospatial Analysis

```python
!pip install folium
```

```python
from sklearn.cluster import KMeans
import folium
import numpy as np
num_clusters = 10

coords = df[['Lat', 'Lon']].values
map_nyc = folium.Map(location=[40.7128, -74.0060], zoom_start=10)

kmeans = KMeans(n_clusters=num_clusters)
df['Cluster'] = kmeans.fit_predict(coords)

for i in range(num_clusters):
    cluster_data = df[df['Cluster'] == i]
    folium.Marker(location=[cluster_data['Lat'].mean(), cluster_data['Lon'].mean()],
                  popup=f'Cluster {i}').add_to(map_nyc)

map_nyc.save("nyc_clusters.html")
map_nyc
```

Clustering algorithms, such as K-Means, group nearby locations into clusters. These clusters help to identify regions where Uber trips are more frequent, which can represent :
- Clustering analysis of Uber trip data helps identify high-demand areas such as airports, nightlife districts, and business hubs

## Base/Driver Analysis

```python
base_perf = df.groupby('Base').size().reset_index(name='RideCount')
base_perf = base_perf.sort_values(by='RideCount', ascending=False)
```

```python
base_perf
```

B02617 has most ride count while B02512 has the least from these 5 Bases.

```python
base_time_perf = df.groupby(['Base', 'Hour']).size().reset_index(name='RideCount')
```

```python
base_time_perf
```

```python

base_time_pivot = base_time_perf.pivot(index='Hour', columns='Base', values='RideCount')
```

```python
base_time_pivot
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))


for base in base_time_pivot.columns:
    plt.plot(base_time_pivot.index, base_time_pivot[base], label=base)


plt.title("Uber Ride Counts by Hour for Each Base", fontsize=16)
plt.xlabel("Hour of the Day", fontsize=12)
plt.ylabel("Number of Rides", fontsize=12)
plt.xticks(range(0, 24))
plt.legend(title='Base', loc='upper right')


plt.tight_layout()
plt.show()
```

- Base B02617 consistently has the highest number of rides across most hours of the day.
- Base B02764 consistently handles the fewest rides, with peak times reaching around 19,000 rides, far below other bases.
- Base B02598 and B02682 also show strong performance, especially during the morning and afternoon peaks.
